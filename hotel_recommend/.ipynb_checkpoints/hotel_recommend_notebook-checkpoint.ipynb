### Exploratory analysis and machine learning

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%pylab inline
```

```python
data_train = pd.read_csv('train.csv', nrows=10)
data_test = pd.read_csv('test.csv', nrows=10)
```

### Correlation

```python
corr = data_train.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(12, 15))

sns.heatmap(corr, mask=mask, vmax=.3, square=True, linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.show()
```

```python
sns.countplot(y='hotel_continent', data=data_train)
sns.plt.title('Continent destination')
plt.show() 
```

```python
sns.countplot(y='hotel_country', data=data_train)
sns.plt.title('Continent destination')
plt.show() 
```

```python
sns.countplot(y='hotel_cluster', data=data_train)
sns.plt.title('Continent destination')
plt.show() 
```
